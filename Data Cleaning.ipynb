```python
df['date']=df.date.apply(lambda x: x.replace('00:00:00',''))
df.head()
```

```python
df['age']=df['yr_built'].apply(lambda x: x if x<1 else 2020-x)
df.head()
```

```python
df['age_since_reno']=df['yr_renovated'].apply(lambda x:x if x<1 else 2020-x)
df.head()
```

```python
df.drop(columns=['statezip','country','street','yr_built','yr_renovated'],inplace=True)
```

```python
df.head()
```

```python
#finding IQR values so as to identify outliers
Q1=df.quantile(0.25)
Q3=df.quantile(0.75)
IQR=Q3-Q1
print(IQR)
```

```python
#printing out the columns having outliers.'True' indicates its presence
print(df<(Q1 - 1.5 * IQR)) | (df>(Q3 + 1.5 * IQR))
```

```python
#removing all outliers and storing them in a new df
df_out=df[~((df<(Q1- 1.5 * IQR)) | (df>(Q3 + 1.5 * IQR))).any(axis=1)]
df_out.shape
```

```python
df_out.head()
```

```python
#certain rows in the price column are zero , so we'll find the mean price and replace those values with it
df['price'].mean()
```

```python
#replacing with mean price
df.price.replace(to_replace=0,value=551962.9884732141,inplace=True)
```
